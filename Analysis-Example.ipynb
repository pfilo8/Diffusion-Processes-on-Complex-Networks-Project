Before running this example please run `run_simulations.py` script to generate data.

```python
import glob
from itertools import product

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
GRAPHS = ['lattice', 'complete']
P = [0.1]
F = [0.1]
Q = [4]
H = [0.09]
```

```python
results = {}

for graph, p, f, q, h in product(GRAPHS, P, F, Q, H):
    current_graph_path = f"{graph}/p-{p}/f-{f}/q-{q}/h-{h}"
    current_graph_simulation = [pd.read_csv(el) for el in glob.glob(f'results/simulations/{current_graph_path}/*.csv')]
    current_graph_simulation = pd.concat(current_graph_simulation)
    results[current_graph_path] = current_graph_simulation
```

```python
results['lattice/p-0.1/f-0.1/q-4/h-0.09']
```

```python
plt.figure(figsize=(8, 5), dpi=300)

sns.lineplot(data = results['lattice/p-0.1/f-0.1/q-4/h-0.09'].mean(axis = 1))
plt.savefig('results/images/example.png')
```
